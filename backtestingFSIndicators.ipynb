```python
class Alphabots(Strategy):

    delta = 0.0007
    volume = 0.5

    def init(self):
        self.short_trade = False
        self.long_trade = False
        self.base_price = 24370.0
        # Initialize indicator functions
        self.rsi = self.I(self.compute_rsi, self.data.Close)
        self.macd_signal = self.I(self.compute_macd, self.data.Close)
        self.bollinger_signal = self.I(self.compute_bollinger, self.data.Close)
        self.adx_signal = self.I(self.compute_adx)

    def compute_rsi(self, prices, period=14):
        rsi = talib.RSI(prices, timeperiod=period)
        return rsi

    def compute_macd(self, prices, fastperiod=12, slowperiod=26, signalperiod=9):
        macd, macdsignal, macdhist = talib.MACD(prices, fastperiod=fastperiod, slowperiod=slowperiod, signalperiod=signalperiod)
        return macd - macdsignal  # Buy when MACD crosses above signal line

    def compute_bollinger(self, prices, period=20, stddev=2):
        upperband, middleband, lowerband = talib.BBANDS(prices, timeperiod=period, nbdevup=stddev, nbdevdn=stddev)
        return upperband, middleband, lowerband

    def compute_adx(self, period=14):
        adx = talib.ADX(self.data.High, self.data.Low, self.data.Close, timeperiod=period)
        return adx

    def generate_signal(self):
      rsi_signal = self.rsi[-1] < 30  # Buy when RSI is below 30 (oversold)
      macd_signal = self.macd_signal[-1] > 0  # Buy when MACD > 0

      # Extract Bollinger Bands values correctly
      upperband = self.bollinger_signal[0][-1]
      middleband = self.bollinger_signal[1][-1]
      lowerband = self.bollinger_signal[2][-1]

      bollinger_signal = self.data.Close[-1] < lowerband  # Buy when price is below the lower Bollinger Band
      adx_signal = self.adx_signal[-1] > 25  # Trend strength (ADX > 25 signals a strong trend)

      # Combine signals (example: majority rule or weighted signals)
      buy_signal = sum([rsi_signal, macd_signal, bollinger_signal, adx_signal]) >= 3  # Buy if 3 out of 4 indicators give a signal
      sell_signal = sum([not rsi_signal, not macd_signal, self.data.Close[-1] > upperband, adx_signal]) >= 3  # Sell if majority signals a sell

      return buy_signal, sell_signal


    def next(self):
      # Handle end of trading day: Close all positions at 15:10
      if self.data.index.time[-1] == time(15, 10):
          print("Closing positions at market close.")
          self.position.close()  # Close any open positions

      # Calculate the threshold prices using delta
      upper_threshold = self.base_price * (1 + self.delta)
      lower_threshold = self.base_price * (1 - self.delta)

      # Generate buy/sell signals based on indicators
      buy_signal, sell_signal = self.generate_signal()

      # Calculate volume based on the combination of signals
      volume = self.volume  # Start with the base volume defined in the class
      if buy_signal:
          volume *= 1 + (sum([buy_signal]) / 4)  # Increase volume based on the strength of buy signals
      if sell_signal:
          volume *= 1 - (sum([sell_signal]) / 4)  # Decrease volume based on the strength of sell signals

      # Ensure volume remains positive
      volume = max(1, volume)  # Ensure volume is at least 1 to avoid errors

      # Check for price crossing upper threshold for buying
      if self.data.Close[-1] > upper_threshold:
          if not self.long_trade:
              print("Executing buy order based on price crossing upper threshold.")
              self.buy(size=volume)  # Use the calculated volume based on indicator signals
              self.long_trade = True
              self.short_trade = False

      # Check for price crossing lower threshold for selling
      elif self.data.Close[-1] < lower_threshold:
          if not self.short_trade:
              print("Executing sell order based on price crossing lower threshold.")
              self.sell(size=volume)  # Use the calculated volume based on indicator signals
              self.short_trade = True
              self.long_trade = False

```

```python
bt = Backtest(day0, Alphabots, cash=200_000, commission=0.00, hedging= True)
stats = bt.run()
```
